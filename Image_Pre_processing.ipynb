**I import exposure library from skimage to performs Gamma Correction**(Also known as Power Law Transform. This function transforms the input image pixelwise according to the equation O = I**gamma after scaling each pixel to the range 0 to 1)on the input image and displayed the bright and dark images after correction the input image:

```
from skimage import exposure
image = imread('/content/drive/MyDrive/example images/dp_beautiful-flowers-profile-dps2014-06-18_12-45-55_1.jpg')
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5,gain=1)

plt.subplot(131), imshow(image)
plt.title('Original Image')
plt.subplot(132),imshow(image_bright)
plt.title('Bright Image')
plt.subplot(133),imshow(image_dark)
plt.title('Dark Image')
plt.show()
```

**skimage.measure.compare_ssim(X, Y[, ...]): Compute the mean structural similarity index between two images.**

**Here,I import compare_ssim from skimage.measure and shown the mean structural similarity index between original and modified images.**

```
from skimage.measure import compare_ssim as ssim
img_original = io.imread('/content/drive/MyDrive/example images/providence-doucet-FjwtL3YSZ9U-unsplash.jpg')
img_modified = io.imread('/content/drive/MyDrive/example images/dp_beautiful-flowers-profile-dps2014-06-18_12-45-55_1.jpg')
img_original=resize(img_original,(300,300))
img_modified=resize(img_modified,(300,300))
ssim_original = ssim(img_original, img_original, data_range=img_original.max() - img_original.min(), multichannel=True)
ssim_different = ssim(img_original, img_modified, data_range=img_modified.max() - img_modified.min(), multichannel=True)
print(ssim_original,ssim_different)
```

**Importing the required libraries**

```
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour
```

The active contour model is a method to fit open or closed splines to lines or edges in an image 1. It works by minimising an energy that is in part defined by the image and part by the splines shape: length and smoothness. The minimization is done implicitly in the shape energy and explicitly in the image energy.

**Through active contour model the segment of the image from the rest of an image by fitting a closed curve to the edges of the face and finding the darkest curve between two fixed points while obeying smoothness considerations.**

```
img=io.imread('/content/drive/MyDrive/example images/dp_beautiful-flowers-profile-dps2014-06-18_12-45-55_1.jpg')
img = rgb2gray(img)

s = np.linspace(0, 2*np.pi, 400)
r = 100 + 100*np.sin(s)
c = 220 + 100*np.cos(s)
init = np.array([r, c]).T

snake = active_contour(gaussian(img, 3),
                       init, alpha=0.015, beta=10, gamma=0.001)

fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(img, cmap=plt.cm.gray)
ax.plot(init[:, 1], init[:, 0], '--r', lw=3)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, img.shape[1], img.shape[0], 0])

plt.show()
```
